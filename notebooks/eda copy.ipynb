```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

pd.set_option("display.max_columns", None)
```

```python
df=pd.read_csv('../data/raw/creditcard.csv')
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.info()
```

```python
## Check missing value
df.isnull().sum()
```

```python
## Class Distribution 
df.Class.value_counts()
```

```python
## Visualize Class Imbalance

sns.countplot(x="Class", data=df)

plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```

```python
## Fraud percentage

fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

print("Fraud transactions:", len(fraud))
print("Normal transactions:", len(normal))

print("Fraud percentage:", round(len(fraud)/len(df)*100, 3), "%")
```

```python
#Transaction Amount Distribution

plt.figure(figsize=(10,5))

sns.histplot(df["Amount"], bins=50)

plt.title("Transaction Amount Distribution")
plt.show()
```

```python
#Fraud vs Non-Fraud Amount

plt.figure(figsize=(10,5))

sns.boxplot(x="Class", y="Amount", data=df)

plt.title("Amount Distribution by Class")
plt.show()
```

```python
# Time Distribution
plt.figure(figsize=(10,5))

sns.histplot(df["Time"], bins=50)

plt.title("Transaction Time Distribution")
plt.show()
```

```python
## Correlation Matrix

corr = df.corr()

plt.figure(figsize=(15,12))

sns.heatmap(corr, cmap="coolwarm")

plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# 
# C

corr_fraud = corr["Class"].sort_values(ascending=False)

corr_fraud.head(10)
```

```python
## Top fraud fetures
top_features = corr_fraud.index[1:6]

for feature in top_features:
    
    plt.figure(figsize=(6,4))
    
    sns.boxplot(x="Class", y=feature, data=df)
    
    plt.title(f"{feature} vs Fraud")
    
    plt.show()
```
